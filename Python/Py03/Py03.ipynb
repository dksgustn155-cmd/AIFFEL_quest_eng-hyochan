```python
import math

def run_calculator():
    print("=== [QUEST 04] 파이썬 계산기 프로그램 ===")

    while True:
        # 1. 첫 번째 정수 입력 및 ValueError 처리
        while True:
            try:
                num1 = int(input("\n첫 번째 정수를 입력하세요: "))
                break # 정수 입력 성공 시 내부 루프 탈출
            except ValueError:
                print("잘못된 입력입니다. 정수를 입력해주세요.")

        # 2. 두 번째 정수 입력 및 ValueError 처리
        while True:
            try:
                num2 = int(input("두 번째 정수를 입력하세요: "))
                break # 정수 입력 성공 시 내부 루프 탈출
            except ValueError:
                print("잘못된 입력입니다. 정수를 입력해주세요.")

        # 3. 연산자 입력 및 재확인 확정
        op_names = {'+': '더하기', '-': '빼기', '*': '곱하기', '/': '나누기', '**': '제곱'}
        while True:
            operator = input(f"연산자를 입력하세요 ({', '.join(op_names.keys())}): ").strip()
            if operator in op_names:
                confirm = input(f"입력하신 연산자가 '{op_names[operator]}({operator})'가 맞습니까? (y/n): ").lower().strip()
                if confirm == 'y':
                    break # 연산자 확정 시 루프 탈출
            else:
                print(f"오류: '{operator}'는 지원하지 않는 연산자입니다.")

        # 4. 실제 계산 및 ZeroDivisionError 처리
        try:
            if operator == '+':
                result = num1 + num2
            elif operator == '-':
                result = num1 - num2
            elif operator == '*':
                result = num1 * num2
            elif operator == '/':
                # 0으로 나누기 예외 상황 체크
                if num2 == 0:
                    raise ZeroDivisionError("0으로 나눌 수 없습니다.")
                result = num1 / num2
            elif operator == '**':
                # math 모듈 사용
                result = math.pow(num1, num2)

            print("-" * 30)
            print(f"결과: {num1} {operator} {num2} = {result}")
            print("-" * 30)

        except ZeroDivisionError as e:
            print(f"⚠️ 계산 오류: {e}")

        # 5. 프로그램 종료 여부 확인
        if input("\n계속 계산하시겠습니까? (종료: n / 계속: 아무키): ").lower() == 'n':
            print("프로그램을 종료합니다.")
            break

if __name__ == "__main__":
    run_calculator()
```
